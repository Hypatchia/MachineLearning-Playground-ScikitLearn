```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
#reading in the data
#The NIFTY 50 index is the National Stock Exchange of India’s benchmark for the Indian equity market
#we’ll be using a subset of the data containing the stock value of only four sectors, i.e., bank, pharma, IT, and FMCG.
nifty_data = pd.read_csv('NIFTY_data_2020.csv',parse_dates=["Date"],index_col='Date')
nifty_data.head()
```

```python
nifty_data.shape
```

```python
#aggregate the data by month-end
#create a dataframe that contains the Mean index resampled by month
nifty_data = nifty_data.rename(columns = {'NIFTY Bank index' : 'Bank index','NIFTY FMCG index':'Company FMCG index'	,'NIFTY IT index': 'company IT index'	,'NIFTY Pharma index'	:'company Pharma index'	})
nifty_data_resample = nifty_data.resample(rule='M').mean()
```

```python
nifty_data_resample =nifty_data_resample.rename(columns = {'NIFTY Bank index' : 'Bank index','NIFTY FMCG index':'Company FMCG index'	,'NIFTY IT index': 'company IT index'	,'NIFTY Pharma index'	:'company Pharma index'	})
```

```python
nifty_data_resample
```

```python
#Visualizing via different plots
#Plotting with Pandas directly
#To plot a graph using pandas, we’ll call the .plot()method on the dataframe
#LINE PLOT
nifty_data.plot(kind = 'scatter',x = 'Company FMCG index',y = 'Bank index',title = 'Scatter Plot for company Index values in 2020',figsize = (10,10))
```

```python
#HISTOGRAMS
nifty_data[['Company FMCG index','Bank index']].plot(kind='hist',figsize=(10,10),bins=30)
```

```python
#BAR PLOTS
nifty_data_resample.plot(kind='bar',figsize=(10,6))
```

```python
#STACKED BAR PLOTS
nifty_data_resample.plot(kind='barh',figsize=(10,6))
```

```python
#PIE CHARTS
nifty_data_resample.index=['Jan','Feb','March','Apr','May','June','July']
nifty_data_resample['Bank index'].plot.pie(legend=False, figsize=(10,6),autopct='%.1f')
```

```python
from bokeh.plotting import figure, output_file, save
figure1= nifty_data.plot_bokeh.line()
output_file(filename="BokehVisualizations1.html", title="Data Visualization Bokeh HTML file")
save(figure1)
```

```python
figure2 =nifty_data.plot_bokeh.scatter(x='Company FMCG index', y='Bank index')
output_file(filename="BokehVisualizations2.html", title="Data Visualization Bokeh HTML file")
save(figure2)
```

```python
figure3=nifty_data[['Company FMCG index','Bank index']].plot_bokeh(kind='hist',bins=30)
output_file(filename="BokehVisualizations3.html", title="Data Visualization Bokeh HTML file")
save(figure3)
```

```python
figure4 =nifty_data_resample.plot_bokeh.bar()
output_file(filename="BokehVisualizations4.html", title="Data Visualization Bokeh HTML file")
save(figure4)
```

```python
figure5 = nifty_data_resample.plot_bokeh(kind='barh',stacked=True)
output_file(filename="BokehVisualizations5.html", title="Data Visualization Bokeh HTML file")
save(figure5)
```

```python
nifty_data_resample.index=['Jan','Feb','March','Apr','May','June','July']
figure6= nifty_data_resample.plot_bokeh.pie(y ='Bank index')
output_file(filename="BokehVisualizations6.html", title="Data Visualization Bokeh HTML file")
save(figure6)
```

```python
figure7=nifty_data_resample.plot_bokeh.pie()
output_file(filename="BokehVisualizations7.html", title="Data Visualization Bokeh HTML file")
save(figure7)
```
